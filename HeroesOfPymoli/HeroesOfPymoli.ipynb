```python
import pandas as pd
```

```python
# Set file path
file = ("Resources/purchase_data.csv")
df = pd.read_csv(file, encoding="latin1")
```

```python
df.head()
```

```python
# PLAYER COUNT

# Total Number of unique players
user_count = len(df.groupby(["SN"]))
pd.DataFrame({'Total Players': [user_count]})
```

```python
# PURCHASING ANALYSIS

# Number of unique items
unique_items =  len(df.groupby(["Item Name"]))

# average purchase price
avg_purchases = df["Price"].mean()
currency_string = "${:,.2f}".format(avg_purchases)

# Total purchases
purchase_count = df["Purchase ID"].count()


# Total Revenue
total_purchases = df["Price"].sum()
total_purchases_curr = "${:,.2f}".format(total_purchases)




df_pa = pd.DataFrame({
    'Number of Unique Items': [unique_items],
    'Average price': [currency_string],
    'Number of Purchases': [purchase_count],
    'Total Revenue': [total_purchases_curr]
})

df_pa

```

```python
df_gender = df.groupby(["SN"])
df_gender.count()
```

```python
# GENDER DEMOGRAPHICS

# Percentage and Count of Male Players 484
male_df = df.loc[df["Gender"]=="Male",:]
male_count = len(male_df.groupby(["SN"]))
male_pct = male_count/user_count
male_pct
# Percentage and Count of Female Players
female_df = df.loc[df["Gender"]=="Female",:]
female_count = len(female_df.groupby(["SN"]))
female_pct = female_count/user_count
female_pct

# Percentage and Count of Other / Non-Disclosed
other_df = df.loc[df["Gender"]=="Other / Non-Disclosed",:]
other_count = len(other_df.groupby(["SN"]))
other_pct = other_count/user_count
other_pct


gender_demo = [{"Total Count": "New Jersey", "Percentage of Players": "NJ"},
                {"Total Count": "New York", "Percentage of Players": "NY"}]


print(male_count, male_pct, female_count, female_pct, other_count, other_pct)
```
